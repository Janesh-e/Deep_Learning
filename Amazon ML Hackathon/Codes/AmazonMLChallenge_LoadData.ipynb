```python
import pandas as pd
import requests
import os
```

```python
# Create a directory to save the images if it doesn't already exist
output_folder = 'amazon_train_images'
os.makedirs(output_folder, exist_ok=True)
```

```python
# Load the dataset into a pandas DataFrame
df = pd.read_csv('Downloads/amazon_ml/student_resource 3/dataset/train.csv')
```

```python
# Function to download and save an image
def download_image(index, url, group_id, folder_path):
    try:
        response = requests.get(url, stream=True)
        response.raise_for_status()  # Check for HTTP request errors
        image_path = os.path.join(folder_path, f'{index}_{group_id}.jpg')  # Save as .jpg or .png based on your needs
        with open(image_path, 'wb') as file:
            file.write(response.content)
        print(f'Successfully downloaded {index}_{group_id}')
    except Exception as e:
        print(f'Failed to download {group_id}: {e}')
```

```python
# Loop through the dataset and download images
for index, row in df.iterrows():
  if(index>=161694):
    image_url = row[0]  # URL is in the first column
    group_id = row[1]  # Group ID is in the second column
    download_image(index, image_url, group_id, output_folder)
```
